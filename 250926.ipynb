### 대입 연산자 (Assignment Operators)

변수에 값을 할당하거나 기존 변수의 값을 변경할 때 사용합니다.

*   `=` : 변수에 값을 할당합니다.
*   `+=` : 왼쪽 변수에 오른쪽 값을 더한 후 결과를 왼쪽 변수에 다시 할당합니다. (`x += y`는 `x = x + y`와 같습니다)
*   `-=` : 왼쪽 변수에서 오른쪽 값을 뺀 후 결과를 왼쪽 변수에 다시 할당합니다. (`x -= y`는 `x = x - y`와 같습니다)
*   `*=` : 왼쪽 변수에 오른쪽 값을 곱한 후 결과를 왼쪽 변수에 다시 할당합니다. (`x *= y`는 `x = x * y`와 같습니다)
*   `/=` : 왼쪽 변수를 오른쪽 값으로 나눈 후 결과를 왼쪽 변수에 다시 할당합니다. (`x /= y`는 `x = x / y`와 같습니다)
*   `%=` : 왼쪽 변수를 오른쪽 값으로 나눈 나머지를 왼쪽 변수에 다시 할당합니다. (`x %= y`는 `x = x % y`와 같습니다)
*   `**=` : 왼쪽 변수를 오른쪽 값만큼 거듭제곱한 후 결과를 왼쪽 변수에 다시 할당합니다. (`x **= y`는 `x = x ** y`와 같습니다)
*   `//=` : 왼쪽 변수를 오른쪽 값으로 나눈 정수 몫을 왼쪽 변수에 다시 할당합니다. (`x //= y`는 `x = x // y`와 같습니다)

### 비교 연산자 (Comparison Operators)

두 값을 비교하여 `True` 또는 `False`를 반환합니다.

*   `==` : 두 값이 같습니다.
*   `!=` : 두 값이 다릅니다.
*   `>` : 왼쪽 값이 오른쪽 값보다 큽니다.
*   `<` : 왼쪽 값이 오른쪽 값보다 작습니다.
*   `>=` : 왼쪽 값이 오른쪽 값보다 크거나 같습니다.
*   `<=` : 왼쪽 값이 오른쪽 값보다 작거나 같습니다.

```python
# 대입 연산자 예시
x = 10
print(f"초기 x 값: {x}")

x += 5  # x = x + 5 와 같음
print(f"x += 5 후 x 값: {x}")

y = 20
y -= 3 # y = y - 3 와 같음
print(f"y -= 3 후 y 값: {y}")

z = 5
z *= 2 # z = z * 2 와 같음
print(f"z *= 2 후 z 값: {z}")

# 비교 연산자 예시
a = 10
b = 5

print(f"a == b: {a == b}") # False
print(f"a != b: {a != b}") # True
print(f"a > b: {a > b}")   # True
print(f"a < b: {a < b}")   # False
print(f"a >= b: {a >= b}") # True
print(f"a <= b: {a <= b}") # False
```

### 논리 연산자 (Logical Operators)

조건문을 결합하거나 논리 값을 반전시킬 때 사용합니다.

중학교때 배운 x축 이상 이하 초과 미만 느낌

*   `and`: 두 조건이 **모두** 참(True)일 때 참(True)을 반환합니다.
*   `or`: 두 조건 중 **하나라도** 참(True)일 때 참(True)을 반환합니다.
*   `not`: 조건의 논리 값을 **반전**시킵니다. (참이면 거짓, 거짓이면 참)

```python
# 논리 연산자 예시

x = 10
y = 5
z = 15

# and 연산자
print(f"x > y and y < z: {x > y and y < z}") # True and True -> True
print(f"x > y and y > z: {x > y and y > z}") # True and False -> False

# or 연산자
print(f"x > y or y > z: {x > y or y > z}")   # True or False -> True
print(f"x < y or y > z: {x < y or y > z}")   # False or False -> False

# not 연산자
print(f"not (x > y): {not (x > y)}")       # not True -> False
print(f"not (x < y): {not (x < y)}")       # not False -> True
```

```python
choice=int(input('1부터 100까지의 짝수 합을 원하시면 0을, 홀수 합을 원하시면 1을 입력해주세요'))

if choice==0:
  print('1부터 100까지의 짝수 합을 선택하셨습니다')
elif choice==1:
  print('1부터 100까지의 홀수 합을 선택하셨습니다')
else:
  print('잘못 선택하셨습니다. 다시 입력해주세요')

even=0
odd=0

for a in range(0,101):
  if a%2==0:
    even=even+a
  else:
    odd=odd+a

if choice==0:
  print(f'1부터 100까지 짝수의 합은 {even} 입니다')
elif choice==1:
  print(f'1부터 100까지 홀수의 합은 {odd} 입니다')
else:
  print()
```

```python
# range(stop) 예시: 0부터 4까지
print("range(5):")
for i in range(5):
  print(i)

# range(start, stop) 예시: 2부터 5까지
print("\nrange(2, 6):")
for i in range(2, 6):
  print(i)

# range(start, stop, step) 예시: 1부터 10까지 2씩 증가
print("\nrange(1, 11, 2):")
for i in range(1, 11, 2):
  print(i)

# range(start, stop, step) 예시: 5부터 1까지 1씩 감소
print("\nrange(5, 0, -1):")
for i in range(5, 0, -1):
  print(i)
```

**→ 2.3 특정 조건이 만족될 때까지 반복: while 문**

**→ 핵심 예제 분석: 5만원 구하기**

```python
sum=0
money=100
day=1
```

```python
# 좌측은 내일, 우측은 오늘 + a
while sum<50000:
  sum=sum+money
  money=money+100
  day=day+1

print('총', day, '일이 걸립니다')
```

```python
# 2일마다 저축하면?
sum=0
money=100
day=1
while sum<50000:
  sum=sum+money
  money=money+100
  day=day+2

print('총', day, '일이 걸립니다')
```

```python
# 숫자를 입력받아서 총합을 구하는 프로그램 작성
# 단, 0을 입력되면 종료되고 최종 총합을 출력한다

total_sum = 0 # 총합을 저장할 변수 초기화

while True: # True가 무한루프를 유발하고 break로 무한루프 끊기
    try:
        num = int(input('숫자를 입력해주세요 (0을 입력하면 종료): '))
        if num == 0:
            break # 0을 입력하면 루프 종료
        total_sum += num # 입력받은 숫자를 총합에 더함
    except ValueError:
        print("잘못된 입력입니다. 숫자를 입력해주세요.")

print(f'입력된 숫자들의 총합은 {total_sum} 입니다.')
```

```python
totalsum=0

while True:
  num=int(input('숫자를 입력해주세요. 0을 입력하면 종료됩니다'))
  if num!=0:
    totalsum=totalsum+num
  else:
    break


print(f'입력된 숫자들의 총합은 {totalsum} 입니다')
```
