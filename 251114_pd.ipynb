** pandas 필수 메소드 **

### 1. 데이터 확인 및 요약 메소드
이 메소드들은 데이터프레임의 구조, 내용, 통계적 특성을 빠르게 파악하는 데 사용됩니다.

→ 이거 앞에 df는 데이터프레임 이름 말하는거

*   **`df.head(n=5)`**: 데이터프레임의 **상위 `n`개 행**을 반환합니다. 기본값은 5개이며, 데이터가 어떻게 생겼는지 가장 먼저 확인하는 데 유용합니다.
*   **`df.tail(n=5)`**: 데이터프레임의 **하위 `n`개 행**을 반환합니다. 데이터의 마지막 부분을 확인할 때 사용합니다.
*   **`df.info()`**: 데이터프레임의 **전반적인 요약 정보**를 출력합니다. 컬럼별 Non-Null 값의 개수, 데이터 타입(dtype), 메모리 사용량 등을 보여주어 결측치와 데이터 타입을 빠르게 파악할 수 있습니다.
*   **`df.describe()`**: 숫자형 컬럼에 대한 **주요 기술 통계량**을 계산하여 반환합니다. 개수, 평균, 표준편차, 최솟값, 최댓값, 사분위수(25%, 50%, 75%) 등을 포함하여 데이터의 분포를 이해하는 데 도움을 줍니다.
*   **`df.shape`**: 데이터프레임의 **행과 열의 개수**(튜플 형태: `(행 수, 열 수)`)를 반환합니다. 데이터의 크기를 확인하는 데 사용됩니다.
*   **`df.columns`**: 데이터프레임의 **모든 컬럼 이름**을 리스트 형태로 반환합니다.
*   **`df.index`**: 데이터프레임의 **인덱스(행 라벨)** 정보를 반환합니다.
*   **`df.dtypes`**: 각 컬럼의 **데이터 타입**을 Series 형태로 반환합니다.

### 2. 데이터 선택 및 접근 메소드
특정 데이터 또는 그룹의 데이터를 선택하는 데 사용되는 매우 중요한 메소드들입니다.

*   **`df.loc[]`**: **레이블(이름)**을 기반으로 데이터를 선택합니다. 행 레이블과 열 레이블을 사용하여 특정 셀, 행, 열 또는 조건에 맞는 데이터를 가져올 수 있습니다. 예를 들어, `df.loc['row_label', 'col_label']` 형태입니다.
*   **`df.iloc[]`**: **정수 위치(순서)**를 기반으로 데이터를 선택합니다. 파이썬 리스트의 인덱싱 및 슬라이싱과 유사하게 동작합니다. 예를 들어, `df.iloc[row_index, col_index]` 형태입니다.
*   **`df['컬럼명']`**: 단일 컬럼을 Series 형태로 선택하거나, `df[['컬럼1', '컬럼2']]`와 같이 여러 컬럼을 DataFrame 형태로 선택합니다.

### 3. 데이터 조작 및 변형 메소드
데이터를 수정하거나 새로운 형태로 만드는 데 사용됩니다.

*   **`df.apply()`**: Series의 각 요소 또는 DataFrame의 각 행/열에 **함수를 적용**합니다. 복잡한 사용자 정의 연산을 데이터에 적용할 때 매우 유용합니다.
*   **`df.groupby()`**: 특정 컬럼의 값을 기준으로 데이터를 **그룹화**합니다. 그룹별 통계 계산(평균, 합계 등)에 주로 사용됩니다.
*   **`df.merge()` / `pd.concat()`**: 여러 개의 데이터프레임을 **병합(join)하거나 연결(concatenate)**합니다. SQL의 JOIN과 유사하게 동작하며, 서로 다른 데이터프레임을 하나로 합칠 때 사용됩니다.
*   **`df.fillna()`**: 데이터프레임 내의 **결측값(NaN)**을 특정 값이나 방법(예: 평균, 이전 값)으로 채웁니다.
*   **`df.dropna()`**: 결측값이 포함된 **행이나 열**을 제거합니다.
*   **`df.drop()`**: 특정 **행이나 열**을 제거합니다.
*   **`df.sort_values()`**: 하나 이상의 컬럼 값을 기준으로 데이터프레임을 **정렬**합니다.
*   **`df.reset_index()`**: 데이터프레임의 인덱스를 **새로운 기본 정수 인덱스**로 재설정하고, 기존 인덱스는 컬럼으로 추가하거나 버립니다.

판다스 데이터 조작 메소드

### 3. 데이터 조작 및 변형 메소드 (데이터 정제 및 변환)

데이터를 수정하거나 새로운 형태로 만드는 데 사용됩니다. 특히 결측치 처리, 컬럼/행 제거, 정렬, 인덱스 재설정 등에 유용합니다.

*   **`df.isnull()`**: 데이터프레임 내의 각 요소가 **결측값(NaN)**인지 여부를 **불리언(True/False) 데이터프레임**으로 반환합니다. 이를 통해 결측치의 위치를 파악할 수 있습니다.
*   **`df.isnull().sum()`**: 각 컬럼별 **결측값의 개수**를 Series 형태로 반환하여, 어떤 컬럼에 결측치가 얼마나 있는지 빠르게 확인할 수 있습니다.
*   **`df.fillna()`**: 데이터프레임 내의 **결측값(NaN)**을 특정 값이나 방법(예: 평균, 이전 값)으로 채웁니다.
*   **`df.dropna()`**: 결측값이 포함된 **행이나 열**을 제거합니다.
*   **`df.drop()`**: 특정 **행이나 열**을 제거합니다.
*   **`df.sort_values()`**: 하나 이상의 컬럼 값을 기준으로 데이터프레임을 **정렬**합니다.
*   **`df.reset_index()`**: 데이터프레임의 인덱스를 **새로운 기본 정수 인덱스**로 재설정하고, 기존 인덱스는 컬럼으로 추가하거나 버립니다.

### `df.isnull()` 예시: `iris_df`에서 결측치 확인

`iris_df`에 결측치가 있는지 확인하고, 각 컬럼별 결측치의 개수를 세어봅니다.

```python
print("--- iris_df의 결측치 확인 (불리언 데이터프레임) ---")
display(iris_df.isnull().head())

print("\n--- 각 컬럼별 결측치 개수 ---")
display(iris_df.isnull().sum())
```

```python
iris_df.shape
```

```python
iris_df.dropna()
```

```python
iris_df.drop_duplicates
```

### `df.drop_duplicates()` 예시: `iris_df`에서 중복 행 제거

`iris_df`의 중복된 행을 제거하고, 제거 전후의 데이터프레임 크기(`shape`)를 비교하여 변경 사항을 확인합니다.

```python
print("--- 중복 제거 전 iris_df의 shape ---")
print(iris_df.shape)

# 'Id' 컬럼은 고유값이므로, 모든 컬럼을 기준으로 중복을 확인하는 것이 더 적절합니다.
# (Id를 제외하고) 나머지 컬럼들을 기준으로 중복을 제거할 수도 있지만, 여기서는 Id를 포함한 모든 컬럼을 기준으로 확인합니다.
df_no_duplicates = iris_df.drop_duplicates()

print("\n--- 중복 제거 후 iris_df (df_no_duplicates)의 shape ---")
print(df_no_duplicates.shape)

print("\n--- 중복 제거 후 데이터프레임 상위 5개 행 ---")
display(df_no_duplicates.head())
```
