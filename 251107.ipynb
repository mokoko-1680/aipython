## NumPy (Numerical Python)

NumPy는 파이썬에서 **수치 계산을 효율적으로 수행하기 위한 핵심 라이브러리**입니다. 특히 다차원 배열 객체인 `ndarray`를 기반으로 하여 대량의 데이터를 빠르고 효율적으로 처리할 수 있게 해줍니다.

### NumPy를 사용하는 이유:
1.  **성능**: 파이썬 리스트에 비해 메모리를 효율적으로 사용하고, C언어로 구현된 내부 루틴을 사용하여 **매우 빠른 연산 속도**를 제공합니다. 이는 특히 대규모 데이터 처리에서 큰 장점입니다.
2.  **다차원 배열(ndarray)**: 벡터, 행렬 등 다양한 형태의 데이터를 쉽게 표현하고 조작할 수 있습니다.
3.  **다양한 수학 함수**: 선형 대수, 푸리에 변환, 난수 생성 등 복잡한 수학 함수들을 내장하고 있어 과학 계산에 필수적입니다.
4.  **브로드캐스팅(Broadcasting)**: 크기가 다른 배열 간에도 연산을 가능하게 하는 강력한 기능을 제공합니다.

주로 데이터 과학, 머신러닝, 과학 컴퓨팅 분야에서 Pandas, Matplotlib, Scikit-learn 등 다른 라이브러리들과 함께 광범위하게 사용됩니다.

### 1. `ndarray` 생성하기

다양한 방법으로 NumPy 배열을 생성할 수 있습니다.

```python
import numpy as np

# 1. 파이썬 리스트로부터 배열 생성
list_data = [1, 2, 3, 4, 5]
numpy_array_from_list = np.array(list_data)
print("파이썬 리스트로부터 생성:\n", numpy_array_from_list)
print("데이터 타입:", numpy_array_from_list.dtype)
print("배열의 형태:", numpy_array_from_list.shape)

# 2. 0부터 n-1까지의 숫자로 배열 생성 (`np.arange()`)
# my_array = np.arange(1_000_0000) 예시에서 이미 사용됨
range_array = np.arange(10) # 0, 1, 2, ..., 9
print("\narange를 이용한 생성:\n", range_array)

# 3. 모든 요소가 0인 배열 생성 (`np.zeros()`)
zeros_array = np.zeros((3, 4)) # 3행 4열의 배열
print("\n모든 요소가 0인 배열:\n", zeros_array)

# 4. 모든 요소가 1인 배열 생성 (`np.ones()`)
ones_array = np.ones((2, 3))
print("\n모든 요소가 1인 배열:\n", ones_array)

# 5. 지정된 시작과 끝 사이를 균등한 간격으로 나눈 배열 생성 (`np.linspace()`)
linspace_array = np.linspace(0, 10, 5) # 0부터 10까지 5개로 나눔
print("\nlinspace를 이용한 생성:\n", linspace_array)
```

### 2. 배열 연산 (Element-wise Operations & Broadcasting)

NumPy 배열은 리스트와 달리 산술 연산자를 사용하면 **요소별(element-wise) 연산**이 수행됩니다. 또한 **브로드캐스팅** 기능을 통해 크기가 다른 배열 간에도 연산이 가능합니다.

```python
import numpy as np

arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

# 요소별 덧셈
sum_array = arr1 + arr2
print("요소별 덧셈:\n", sum_array)

# 요소별 곱셈 (스칼라 곱셈 포함)
product_array = arr1 * 3
print("스칼라 곱셈:\n", product_array)

# 브로드캐스팅 예시: 배열과 스칼라 연산
broadcast_array = np.array([[1, 2], [3, 4]])
broadcast_result = broadcast_array + 10
print("\n브로드캐스팅 연산:\n", broadcast_result)

# 두 배열 간 브로드캐스팅 (크기가 맞지 않아도 규칙에 따라 확장되어 연산)
# 예를 들어, (3,) 배열과 (3, 1) 배열 간 연산 등
```

### 3. 배열 인덱싱 및 슬라이싱

NumPy 배열도 파이썬 리스트처럼 인덱싱 및 슬라이싱이 가능하며, 다차원 배열에서는 더욱 강력한 기능을 제공합니다.

```python
import numpy as np

arr = np.array([[10, 20, 30], [40, 50, 60], [70, 80, 90]])
print("원본 배열:\n", arr)

# 단일 요소 접근
print("\n첫 번째 행, 두 번째 열 요소:", arr[0, 1]) # 20

# 행 전체 슬라이싱
print("\n두 번째 행 전체:\n", arr[1, :]) # [40 50 60]

# 열 전체 슬라이싱
print("\n세 번째 열 전체:\n", arr[:, 2]) # [30 60 90]

# 부분 배열 슬라이싱
print("\n부분 배열 (첫 두 행, 첫 두 열):\n", arr[:2, :2]) # [[10 20], [40 50]]

# 조건부 인덱싱 (Boolean Indexing)
# 50보다 큰 요소만 선택
condition = arr > 50
print("\n50보다 큰 요소들의 불리언 배열:\n", condition)
print("50보다 큰 요소들:\n", arr[condition]) # [60 70 80 90]
```

### 4. 배열 형태 변경 (Reshaping)

배열의 데이터는 그대로 유지한 채 형태(차원)를 변경할 수 있습니다.

```python
import numpy as np

arr = np.arange(12) # 0부터 11까지의 숫자
print("원본 1차원 배열:\n", arr)

# 3행 4열의 2차원 배열로 변경
reshaped_arr = arr.reshape(3, 4)
print("\n3x4로 형태 변경:\n", reshaped_arr)

# -1을 사용하면 NumPy가 적절한 차원수를 자동으로 계산합니다.
reshaped_arr_auto = arr.reshape(2, -1) # 2행으로 만들고 열 수는 자동 계산 (6열)
print("\n2x자동으로 형태 변경:\n", reshaped_arr_auto)

# 1차원 배열로 평탄화 (flatten)
flattened_arr = reshaped_arr.flatten()
print("\n평탄화된 배열:\n", flattened_arr)
```

### 5. 통계 및 수학 함수

NumPy는 배열에 적용할 수 있는 다양한 통계 및 수학 함수를 제공합니다.

```python
import numpy as np

arr = np.array([[1, 2, 3], [4, 5, 6]])
print("원본 배열:\n", arr)

# 합계
print("\n전체 요소의 합계:", np.sum(arr))

# 평균
print("전체 요소의 평균:", np.mean(arr))

# 최댓값
print("전체 요소의 최댓값:", np.max(arr))

# 축(axis)을 지정하여 연산 (예: 행별 또는 열별)
print("\n각 열의 합계 (axis=0):", np.sum(arr, axis=0)) # [1+4, 2+5, 3+6] = [5 7 9]
print("각 행의 평균 (axis=1):", np.mean(arr, axis=1)) # [(1+2+3)/3, (4+5+6)/3] = [2. 5.]

# 표준 편차
print("전체 요소의 표준 편차:", np.std(arr))
```

## `np.arange()` 자세히 알아보기

`np.arange()` 함수는 NumPy 라이브러리에서 **지정된 간격으로 균일하게 분포된 값을 가지는 배열을 생성**하는 데 사용됩니다. 파이썬의 내장 `range()` 함수와 개념은 유사하지만, `np.arange()`는 항상 NumPy의 `ndarray` (다차원 배열) 객체를 반환한다는 중요한 차이점이 있습니다.

### `np.arange()`의 기본 구문:
`numpy.arange([start,] stop[, step,], dtype=None)`

*   `start` (선택 사항): 수열의 시작 값입니다. 기본값은 0입니다. (포함)
*   `stop`: 수열의 끝 값입니다. 이 값은 수열에 포함되지 않습니다. (미포함)
*   `step` (선택 사항): 수열 내 값들의 간격입니다. 기본값은 1입니다. 실수(float) 값도 가능합니다.
*   `dtype` (선택 사항): 생성될 배열의 데이터 타입을 명시합니다. 지정하지 않으면 입력 값에 따라 자동으로 추론됩니다.

### 주요 특징 및 사용 사례:

1.  **정수 수열 생성**: 가장 기본적인 사용법으로, `stop` 값만 지정하면 0부터 `stop-1`까지의 정수 배열을 만듭니다.
2.  **시작/끝/간격 지정**: `start`, `stop`, `step`을 모두 지정하여 원하는 범위와 간격의 수열을 생성할 수 있습니다.
3.  **실수 간격 지원**: `step` 인자에 실수를 사용하여 부동소수점 값으로 구성된 수열도 쉽게 만들 수 있습니다. (파이썬의 `range()`는 정수만 가능)
4.  **효율적인 배열 생성**: 대량의 숫자 데이터를 빠르게 생성하고 처리해야 할 때 파이썬 리스트의 `range()`보다 훨씬 효율적입니다.
5.  **메모리 효율성**: NumPy 배열은 파이썬 리스트보다 메모리를 적게 사용하며, 데이터 과학 및 머신러닝에서 광범위하게 사용됩니다.

```python
import numpy as np

print("--- np.arange() 예시 ---")

# 1. stop 값만 지정 (0부터 stop-1까지)
# np.arange(5) -> [0, 1, 2, 3, 4]
print("예시 1: np.arange(5)\n", np.arange(5))

# 2. start와 stop 값 지정 (start부터 stop-1까지)
# np.arange(2, 7) -> [2, 3, 4, 5, 6]
print("\n예시 2: np.arange(2, 7)\n", np.arange(2, 7))

# 3. start, stop, step 값 지정 (start부터 step 간격으로 stop-1까지)
# np.arange(1, 10, 2) -> [1, 3, 5, 7, 9]
print("\n예시 3: np.arange(1, 10, 2)\n", np.arange(1, 10, 2))

# 4. 실수 간격(step) 지정
# np.arange(0.0, 1.0, 0.2) -> [0.0, 0.2, 0.4, 0.6, 0.8]
print("\n예시 4: np.arange(0.0, 1.0, 0.2)\n", np.arange(0.0, 1.0, 0.2))

# 5. 음수 간격(step) 지정 (역순으로 배열 생성)
# np.arange(10, 0, -2) -> [10, 8, 6, 4, 2]
print("\n예시 5: np.arange(10, 0, -2)\n", np.arange(10, 0, -2))

# 6. 데이터 타입(dtype) 명시
# np.arange(3, dtype=float) -> [0., 1., 2.]
print("\n예시 6: np.arange(3, dtype=float)\n", np.arange(3, dtype=float))

# 7. 큰 숫자 생성 (가독성을 위한 1_000_000)
# np.arange(1_000_000)는 0부터 999,999까지의 백만 개의 정수를 포함하는 배열 생성
# 이 예시는 너무 커서 출력을 생략하고 배열의 길이만 확인합니다.
large_array = np.arange(1_000_000)
print(f"\n예시 7: np.arange(1_000_000)으로 생성된 배열의 길이: {len(large_array)}")
print(f"예시 7: np.arange(1_000_000)으로 생성된 배열의 처음 5개 요소: {large_array[:5]}")
```

[7] 실전예제

```python
import numpy as np

# 비행 거리 배열 (단위: km)
distances = np.array([1, 2, 3, 5, 8])

# 각 비행 시 예상 풍속 배열 (단위: m/s)
windspeeds = np.array([0, 5, 10, 7, 12])
```

```python
# 기본 배터리 소모율 (단위: %/km)
base_usage_rate = 10

# 풍속에 따른 추가 소모율 (단위: %/km)
additional_usage_rate = windspeeds

# 각 시나리오의 최종 소모율 (%/km)
total_usage_rate = base_usage_rate + additional_usage_rate

# 최종 소모량 (%) = 거리(km) * 소모율(%/km)
battery_consumption = distances * total_usage_rate
```

```python
print("비행 거리 (km):", distances)
print("풍속 (m/s):", windspeeds)
print("총 소모율 (%/km):", total_usage_rate)
print("예상 배터리 소모량 (%):", battery_consumption)
```
