```python
fruits=['사과','바나나','딸기']
counts=[10,5,7]
colors=['green', 'yellow','red']
```

```python
plt.bar(fruits,counts,color=colors)
plt.title('과일 판매량',fontsize=16)
plt.xlabel('과일종류', fontsize=14)
plt.ylabel('판매수량', fontsize=14)

plt.show()
```

```python
selling_fruits=plt.bar(fruits,counts,color=colors)
plt.grid(selling_fruits)
plt.show(selling_fruits)
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
# iris 데이터 로딩
# seaborn안에 연습용 데이터셋 iris가 있음
iris=sns.load_dataset('iris')
```

```python
# 크기 조절
plt.figure(figsize=(10,7))
```

```python
# x축 지정, y축 지정
# hue는 데이터 범주, s는 점 크기 조절, alpha는 점의 투명도 -> 이건 필수가 아님
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris, s=100, alpha=0.8)
```
